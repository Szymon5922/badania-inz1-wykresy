```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
```

zdefiniowanie połączenia z plikiem bazodanowym i utworzenie potrzebnych dataframes

```python
conn = sqlite3.connect('data/badania_inz1.db')
query = "SELECT * FROM t_czerwiec"
df_czerwiec = pd.read_sql_query(query,conn)
query = "SELECT * FROM t_lipiec"
df_lipiec = pd.read_sql_query(query,conn)
query = "SELECT * FROM t_dodatkowa"
df_dodatkowa = pd.read_sql_query(query,conn)
```

```python

# Grupowanie danych po 'Wilgotnosc_gleby' i obliczanie średniej, min i max 'Wskaznik_zawartosci_chlorofilu'
grouped_data = df_czerwiec.groupby('Wilgotnosc_gleby')['Wskaznik_zawartosci_chlorofilu'].agg(['mean', 'min', 'max'])

# Tworzenie wykresu
plt.figure(figsize=(20, 10))

# Wykres linii reprezentującej średnią wartość 'Wskaznik_zawartosci_chlorofilu'
plt.plot(grouped_data.index, grouped_data['mean'], color='blue', label='Średnia wartość dla wilgotności')

# Wykres kropek dla średniej wartości 'Wskaznik_zawartosci_chlorofilu'
plt.scatter(grouped_data.index, grouped_data['mean'], color='blue')

# Wypełnienie obszaru między minimalną a maksymalną wartością 'Wskaznik_zawartosci_chlorofilu'
plt.fill_between(grouped_data.index, grouped_data['min'], grouped_data['max'], color='skyblue', alpha=0.5, label='Min-Max zakres')

plt.axhline(df_czerwiec['Wskaznik_zawartosci_chlorofilu'].mean(),color='green',label='Średnia wartość ogółem')

# Zaznaczenie interesujących wartości osi x
interesujace_wilgotnosci = [1, 3, 4, 5, 6, 8]  # Tu podaj interesujące wartości 'Wilgotnosc_gleby'
plt.xticks(interesujace_wilgotnosci)
plt.yticks(np.arange(0,18,0.5))
# Dodanie siatki
plt.grid(True, linestyle='--', linewidth=0.5)

# Dodanie legendy
plt.legend()

# Dodanie tytułów i opisów osi
plt.xlabel('Wilgotność gleby zmierzna testerem podłoża')
plt.ylabel('Wskaźnik zawartości chlorofilu')
plt.title('Zależność zawartości chlorofilu od wilgotności gleby (Czerwiec)')

# Pokazanie wykresu
plt.show()
```

```python


# Grupowanie danych po 'Wilgotnosc_gleby' i obliczanie średniej i odchylenia standardowego 'Wskaznik_zawartosci_chlorofilu'
grouped_data = df_czerwiec.groupby('Wilgotnosc_gleby')['Wskaznik_zawartosci_chlorofilu'].agg(['mean', 'std'])

# Pobranie wartości 'Wilgotnosc_gleby' jako indeksów oraz 'mean' i 'std' jako wartości dla wykresu błędów
wilgotnosc_gleby = grouped_data.index
srednia_wartosc = grouped_data['mean']
odchylenie_standardowe = grouped_data['std']

# Ustawienie większego rozmiaru wykresu
plt.figure(figsize=(20, 10))
interesujace_wilgotnosci = [1, 3, 4, 5, 6, 8]  # Tu podaj interesujące wartości 'Wilgotnosc_gleby'
plt.xticks(interesujace_wilgotnosci)
plt.yticks(np.arange(0,18,0.5))
# Tworzenie wykresu z paskami błędów
plt.errorbar(wilgotnosc_gleby, srednia_wartosc, yerr=odchylenie_standardowe, fmt='o', color='blue', ecolor='red', capsize=5, label='Wskaznik_zawartosci_chlorofilu')

# Dodanie siatki i legendy
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()

# Dodanie tytułów i opisów osi
plt.xlabel('Wilgotność gleby zmierzona testerem podłoża')
plt.ylabel('Wskaźnik zawartości chlorofilu')
plt.title('Zależność zawartości chlorofilu od wilgotności gleby (Czerwiec)')

# Pokazanie wykresu
plt.show()
```

```python
grouped_data = df_czerwiec.groupby('Umiejscowienie_proby')['Wskaznik_zawartosci_chlorofilu'].apply(list)

# Konwersja do listy list wartości
data_for_violin = grouped_data.values.tolist()

# Tworzenie wykresu skrzypcowego
plt.figure(figsize=(14, 8))
plt.violinplot(data_for_violin)
plt.xlabel('Umiejscowienie_proby')
plt.ylabel('Wskaznik_zawartosci_chlorofilu')
plt.title('Wykres skrzypcowy Wskaznik_zawartosci_chlorofilu dla Umiejscowienie_proby')

# Dodanie kernel density plot (rozkładu gęstości)
for i, data in enumerate(data_for_violin, start=1):
    positions = np.random.normal(i, 0.04, size=len(data))  # Aby uniknąć nakładania się wykresów
    plt.scatter(positions, data, alpha=0.5, color='black')

# Plotting the line chart for average values
mean_values = [np.mean(values) for values in data_for_violin]
plt.plot(range(1, len(mean_values) + 1), mean_values, marker='o', linestyle='-', color='red', label='Średnia wartość')

plt.grid(True)
plt.xticks(np.arange(1, len(grouped_data) + 1), grouped_data.index)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare data for violin plot
data_for_violin = chlorofilPerWilgotnosc.values.tolist()
labels = chlorofilPerWilgotnosc.index.tolist()  # Wykorzystujemy indeksy jako etykiety na osi X

# Create violin plot
plt.figure(figsize=(10, 6))
violins = plt.violinplot(data_for_violin, positions=range(1, len(data_for_violin) + 1), widths=0.9)

# Add points for individual values with some random jitter
for i, points in enumerate(data_for_violin):
    y = np.array(points)
    x = np.random.normal(i + 1, 0.05, size=len(points))
    plt.scatter(x, y, color='black', s=10, alpha=0.5)

# Plotting the line chart for average values
mean_values = [np.mean(values) for values in data_for_violin]
plt.plot(range(1, len(mean_values) + 1), mean_values, marker='o', linestyle='-', color='red', label='Średnia wartość')

# Set labels and title
plt.xlabel('Wilgotnosc_gleby')
plt.ylabel('Wskaznik_zawartosci_chlorofilu')
plt.title('Wykres skrzypcowy Zawartosci chlorofilu w zaleznosci od Wilgotnosc_gleby')

# Set x-axis ticks and labels
plt.xticks(range(1, len(labels) + 1), labels)

plt.grid(True)

# Legend for the points and line chart
plt.scatter([], [], color='black', label='Wskaznik_zawartosci_chlorofilu', alpha=0.5)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Przygotowanie danych dla wykresu skrzypcowego
naslonecznienie_values = sorted(df_czerwiec['Naslonecznienie'].unique())
data = [df_czerwiec[df_czerwiec['Naslonecznienie'] == naslonecznienie]['Wskaznik_zawartosci_chlorofilu'].values 
        for naslonecznienie in naslonecznienie_values]

# Rysowanie skrzypcowego wykresu wiolinowego
plt.figure(figsize=(12, 6))
violins = plt.violinplot(data, showmedians=True)

# Dodawanie wykresu liniowego dla średnich wartości 'Wskaznik_zawartosci_chlorofilu' dla poszczególnych wartości 'Naslonecznienie'
mean_values = df_czerwiec.groupby('Naslonecznienie')['Wskaznik_zawartosci_chlorofilu'].mean()
plt.plot(range(1, len(naslonecznienie_values) + 1), mean_values.values, marker='o', linestyle='-', color='red', label='Średnia wartość')

# Dodanie punktów do wykresu dla poszczególnych wartości z lekkim przesunięciem
for i, naslonecznienie in enumerate(naslonecznienie_values):
    values = df_czerwiec[df_czerwiec['Naslonecznienie'] == naslonecznienie]['Wskaznik_zawartosci_chlorofilu']
    x = np.full(len(values), i + 1) + np.random.normal(0, 0.05, len(values))  # Dodanie szumu do wartości x
    plt.scatter(x, values, color='black', alpha=0.5)

# Ustawienie etykiet osi X
plt.xticks(range(1, len(naslonecznienie_values) + 1), naslonecznienie_values)

# Etykiety osi i tytuł
plt.xlabel('Naslonecznienie')
plt.ylabel('Wskaznik_zawartosci_chlorofilu')
plt.title('Wykres skrzypcowy Zawartosci chlorofilu w zależności od Naslonecznienie')

plt.legend()  # Dodanie legendy
plt.grid(True)
plt.show()
```

```python
mean_chlorofil_per_wilgotnosc_czerwiec = df_czerwiec.groupby('Wilgotnosc_gleby')['Wskaznik_zawartosci_chlorofilu'].mean()
print(mean_chlorofil_per_wilgotnosc_czerwiec)
```

```python
mean_chlorofil_per_wilgotnosc_lipiec = df_lipiec.groupby('Wilgotnosc_gleby')['Wskaznik_zawartosci_chlorofilu'].mean()
print(mean_chlorofil_per_wilgotnosc_lipiec)
```

```python
mean_chlorofil_per_wilgotnosc_dodatkowa = df_dodatkowa.groupby('Wilgotnosc_gleby')['Wskaznik_zawartosci_chlorofilu'].mean()
print(mean_chlorofil_per_wilgotnosc_dodatkowa)
```

```python
plt.figure(figsize=(10, 6))

# Łączenie indeksów unikalnych wartości wilgotności gleby z obu miesięcy
unique_indexes = sorted(set(mean_chlorofil_per_wilgotnosc_czerwiec.index) | set(mean_chlorofil_per_wilgotnosc_lipiec.index))
bar_width = 0.35
index = np.arange(len(unique_indexes))

# Pobranie wartości chlorofilu dla czerwca i lipca, zachowując równą długość dla obu zestawów danych
chlorofil_czerwiec = [mean_chlorofil_per_wilgotnosc_czerwiec.get(idx, 0) for idx in unique_indexes]
chlorofil_lipiec = [mean_chlorofil_per_wilgotnosc_lipiec.get(idx, 0) for idx in unique_indexes]

plt.bar(index, chlorofil_czerwiec, bar_width, label='Czerwiec')
plt.bar(index + bar_width, chlorofil_lipiec, bar_width, label='Lipiec')

plt.xlabel('Poziom wilgotności gleby')
plt.ylabel('Średnia wartość chlorofilu')
plt.title('Średnia wartość chlorofilu dla poziomów wilgotności gleby w czerwcu i lipcu')
plt.xticks(index + bar_width / 2, unique_indexes)
plt.legend()

plt.tight_layout()
plt.show()
```
