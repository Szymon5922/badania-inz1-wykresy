```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
```

```python
conn = sqlite3.connect('data/badania_inz1.db')
#query = "SELECT * FROM t_czerwiec"
#df_czerwiec = pd.read_sql_query(query,conn)
#query = "SELECT * FROM t_lipiec"
#df_lipiec = pd.read_sql_query(query,conn)
query = "SELECT * FROM t_dodatkowa"
df_dodatkowa = pd.read_sql_query(query,conn)
```

```python
df_dodatkowa.dtypes
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grupowanie danych dla dodatkowej kategorii
grouped_data_dodatkowa = df_dodatkowa.groupby('Umiejscowienie_proby')['Wskaznik_zawartosci_chlorofilu'].apply(list)
data_for_violin_dodatkowa = grouped_data_dodatkowa.values.tolist()

plt.style.use('seaborn')
plt.figure(figsize=(8, 5))
plt.violinplot(data_for_violin_dodatkowa)
plt.xlabel('Umiejscowienie prby')
plt.ylabel('Wskanik zawartoci chlorofilu')
plt.title('Wskanik zawartoci chlorofilu wzgldem umiejscowienia prby. Oddzia: 733-c Data: 7/27/2023', fontsize=10)

for i, data in enumerate(data_for_violin_dodatkowa, start=1):
    positions = np.random.normal(i, 0.04, size=len(data))  
    plt.scatter(positions, data, alpha=0.5, color='black', label='Poszczeglne odczyty' if i == 1 else None)

# Dodaj czerwone punkty oznaczajce redni warto
mean_values_dodatkowa = [np.mean(values) for values in data_for_violin_dodatkowa]
plt.scatter(np.arange(1, len(grouped_data_dodatkowa) + 1), mean_values_dodatkowa, marker='o', color='red', label='rednia warto')

# Dodaj adnotacje obok punktw oznaczajcych redni warto
for i, mean_value in enumerate(mean_values_dodatkowa, start=1):
    plt.text(i + 0.3, mean_value, f'{mean_value:.2f}', ha='center', va='bottom', color='red')

plt.yticks([i for i in range(0, 17, 1)])
plt.grid(True)
plt.xticks(np.arange(1, len(grouped_data_dodatkowa) + 1), grouped_data_dodatkowa.index)
plt.legend()
plt.show()
```

```python
chlorofilPerWilgotnosc = df_dodatkowa.groupby('Wilgotnosc_gleby')['Wskaznik_zawartosci_chlorofilu'].apply(list)
print(chlorofilPerWilgotnosc)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Twoje dane
data_for_violin = chlorofilPerWilgotnosc.values.tolist()
labels = chlorofilPerWilgotnosc.index.tolist() 

plt.style.use('seaborn')

plt.figure(figsize=(10, 6))
violins = plt.violinplot(data_for_violin, positions=range(1, len(data_for_violin) + 1), widths=0.9)

for i, points in enumerate(data_for_violin):
    y = np.array(points)
    x = np.random.normal(i + 1, 0.05, size=len(points))
    plt.scatter(x, y, color='black', s=10, alpha=0.5)

mean_values = [np.mean(values) for values in data_for_violin]
plt.scatter(range(1, len(mean_values) + 1), mean_values, marker='o', linestyle='-', color='red', label='rednia warto')

# Dodaj liczb obok kadej kropki oznaczajcej warto redni
for i, mean_value in enumerate(mean_values):
    plt.text(i + 1.1, mean_value, f'{mean_value:.2f}', ha='center', va='bottom', color='red')

plt.xlabel('Wilgotno gleby zmierzona testerem podoa')
plt.ylabel('Wskanik zawartoci chlorofilu')
plt.title('Wskanik zawartoci chlorofilu w zalenoci od wilgotnoci gleby. Oddzia: 733-c Data: 7/27/2023',fontsize=10)

plt.xticks(range(1, len(labels) + 1), labels)
plt.yticks([i for i in range(0, 30, 1)])

plt.grid(True)

plt.scatter([], [], color='black', label='Poszczeglne odczyty', alpha=0.5)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

naslonecznienie_values = sorted(df_dodatkowa['Naslonecznienie'].unique())
data = [df_dodatkowa[df_dodatkowa['Naslonecznienie'] == naslonecznienie]['Wskaznik_zawartosci_chlorofilu'].values 
        for naslonecznienie in naslonecznienie_values]

# Etykiety dla osi x
additional_labels = ["\n3", "\n2", "\n4", "\n1"]

plt.style.use('seaborn')
plt.figure(figsize=(10, 6))

# Rysuj wykresy wiolinowe
violins = plt.violinplot(data)

# Rysuj kropki oznaczajce poszczeglne odczyty wartoci chlorofilu
for i, naslonecznienie in enumerate(naslonecznienie_values):
    values = df_dodatkowa[df_dodatkowa['Naslonecznienie'] == naslonecznienie]['Wskaznik_zawartosci_chlorofilu']
    x = np.full(len(values), i + 1) + np.random.normal(0, 0.05, len(values))  
    plt.scatter(x, values, color='black', alpha=0.5, label='Poszczeglne odczyty' if i == 1 else None)

# Rysuj kropki ze redni wartoci chlorofilu
mean_values = df_dodatkowa.groupby('Naslonecznienie')['Wskaznik_zawartosci_chlorofilu'].mean()
plt.scatter(range(1, len(naslonecznienie_values) + 1), mean_values.values, marker='o', color='red', label='rednia warto')

# Dodaj adnotacje obok kropek oznaczajcych redni warto
for i, mean_value in enumerate(mean_values):
    plt.annotate(f'{mean_value:.2f}', (i + 1.3, mean_value), textcoords="offset points", xytext=(0, 5), ha='center', va='bottom', color='red')

# Dodaj dodatkowe etykiety do osi x
plt.xticks(range(1, len(naslonecznienie_values) + 1), [f"{value} {label}" for value, label in zip(naslonecznienie_values, additional_labels)])
plt.yticks([i for i in range(0, 29, 1)])
plt.xlabel('Nasonecznienie w PPFD [mol/m2/s], umiejscowienie prby')
plt.ylabel('Wskanik zawartoci chlorofilu')
plt.title('Wskanik zawartoci chlorofilu w zalenoci od nasonecznienia i umiejscowienia prby.\nOddzia: 733-c Data: 7/27/2023', fontsize=10)

plt.legend()
plt.grid(True)
plt.show()
```
